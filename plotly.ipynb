```python
#scatterplot using plotly go

temp_df= gap[gap['year']==2007]
```

```python
trace1 = go.Scatter(
    x=temp_df['lifeExp'],
    y=temp_df['gdpPercap'],
    mode='markers'
)

trace2 = go.Scatter(x=[0,1,2],
                    y=[0,90,30000],
                    mode='lines')


data = [trace1,trace2]

layout = go.Layout(
    title='Life Expectancy vs GDP per Capita (2007)',
    xaxis=dict(title='Life Expectancy'),
    yaxis=dict(title='GDP per Capita')
)

fig = go.Figure(data=data, layout=layout)
fig.show()
```

```python
#plot life exp and gdp scatter plot ->content as color -> pop as size -> hover name -> range_x/range_y ->log_x/log_y

px.scatter(temp_df,x='lifeExp',y='gdpPercap',
           color='continent',
           size='pop',size_max=100,
           hover_name='country')
```

```python
#plot animation of the above curve on the basis of year
px.scatter(gap,x='lifeExp',y='gdpPercap',
           color='continent',
           size='pop',size_max=100,
           hover_name='country',
           animation_frame='year',
           animation_group='country')
```

```python
#3d surface plot
# can not be created using plotly express
#we will use plotly graph object -> go

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)

#z=xx**2 +yy**2
z=np.sin(xx)+np.tan(yy)

trace=go.Surface(x=xx,y=yy,z=z)

data=[trace]

layout=go.Layout(title='3d Surface Plot')

fig=go.Figure(data=[go.Surface(x=xx,y=yy,z=z)])

fig.show()
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)

#z=xx**2 +yy**2
#z=np.sin(xx)+np.tan(yy)
z=np.sqrt(xx**2+yy**2)

trace=go.Surface(x=xx,y=yy,z=z)

data=[trace]

layout=go.Layout(title='3d Surface Plot')

fig=go.Figure(data=[go.Surface(x=xx,y=yy,z=z)])

fig.show()
```

```python
#contour plot
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
xx, yy = np.meshgrid(x, y)

#z = xx**2 + yy**2
z=np.sin(xx)+np.tan(yy)

fig = go.Figure(
    data=[
        go.Contour(
            x=x,        # 1D array
            y=y,        # 1D array
            z=z         # 2D array
        )
    ]
)

fig.update_layout(title='Contour Plot')

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# create subplot grid
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=[
        'Scatter (go)',
        'Histogram (go)',
        'Scatter (markers)',
        'Histogram'
    ]
)

# row 1, col 1 â†’ scatter
fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[4, 5, 6], mode='lines+markers'),
    row=1,
    col=1
)

# row 1, col 2 â†’ histogram
fig.add_trace(
    go.Histogram(x=[1, 2, 3, 4, 5, 6, 10, 40, 24, 50, 42, 53]),
    row=1,
    col=2
)

# row 2, col 1 â†’ scatter (markers)
fig.add_trace(
    go.Scatter(x=[1, 2, 3], y=[4, 5, 6], mode='markers'),
    row=2,
    col=1
)

# row 2, col 2 â†’ histogram
fig.add_trace(
    go.Histogram(x=[1, 2, 5, 3, 2, 53, 43, 234, 64, 7, 31, 86]),
    row=2,
    col=2
)

fig.update_layout(title='Subplot Demo')

fig.show()
```
